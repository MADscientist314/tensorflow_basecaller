```
def remove_insignificant_data(arr, delta, min_len):
  good_arr = []
  temp_arr = []

  for i in range(len(arr)):
    if i == 0:
      temp_arr.append(arr[i])
      continue

    if(abs(arr[i-1] - arr[i]) <= delta):
      temp_arr.append(arr[i])
      continue

    if(len(temp_arr) >= min_len):
      good_arr.append(temp_arr)

    temp_arr = [arr[i]]

  if(len(temp_arr) >= min_len):
    good_arr.append(temp_arr)
  
  return good_arr
```

```
ret_arr_sig = np.array(remove_insignificant_data(np.round(sig[0],4), 0.005, 3))
ret_arr_reads = remove_insignificant_data(reads[10], 0.5, 1)
print(len(sig[1]))
print(len(reads[1]))
print(len(ret_arr_sig))
print(len(ret_arr_reads))
print(ret_arr_sig[:10])
print(ret_arr_reads[:10])
print(reads[0][:50])
```

```
flat_arr_sig = []
flat_arr_read = []

for arr in ret_arr_sig:
  flat_arr_sig.extend(arr)
  
for arr in ret_arr_reads:
  flat_arr_read.extend(arr)

flat_arr_sig = np.array(flat_arr_sig)
flat_arr_read = np.array(flat_arr_read)
```

```
km = KMeans(4)
km.fit(flat_arr_sig.reshape(-1,1))
print(km.cluster_centers_)
```

```
predict_arr = []
for i in flat_arr_sig:
  predict_arr.extend(km.predict(np.array(i).reshape(-1,1)))

```
